## Pickling the data of selected Twitter Accounts

```python
import tweepy
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
import pickle

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)

public_tweets = api.user_timeline('BCCI')
public_tweets_1 = api.user_timeline('BeingSalmanKhan')
public_tweets_2 = api.user_timeline('aamir_khan')

with open('BCCI.pkl', 'wb') as f:
    pickle.dump(public_tweets, f)
with open('BeingSalmanKhan.pkl', 'wb') as f:
    pickle.dump(public_tweets_1, f)
    
with open('aamir_khan.pkl', 'wb') as f:
    pickle.dump(public_tweets_2, f)
```

## Loading the Pickled data

```python
import pickle

#first account
with open('BCCI.pkl', 'rb') as f:
    public_tweets = pickle.load(f)
print(public_tweets)
#second account
with open('BeingSalmanKhan.pkl', 'rb') as f:
    public_tweets_1 = pickle.load(f)

print(public_tweets_1)
#third account
with open('aamir_khan.pkl', 'rb') as f:
    public_tweets_2 = pickle.load(f)

print(public_tweets_2)
```

# Retweet Count Calculation of a Tweet

# Analysis Steps
1. Retweet refers to reposting or forwarding a message posted by another user.
2. Here we have pulled out a tweet of our selected users and calculated its retweet count.
3. We used retweet_count method for getting this count.
4. Plotted the retweet count in a bar chart and results are shown below.

```python
# Calculating retweet count of a particular tweet

bcci_retweet= public_tweets[4].retweet_count
salman_retweet = public_tweets_1[4].retweet_count
aamir_retweet = public_tweets_2[4].retweet_count

print('Retweet Count of BCCI:',bcci_retweet)
print('Retweet Count of Salman Khan:', salman_retweet)
print('Retweet Count of Aamir Khan:',aamir_retweet)
```

# Plotting the Retweet Count of a Tweet using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
objs = ('bcci_retweet','salman_retweet','aamir_retweet')
y_position = np.arange(len(objs))
perform = [bcci_retweet, salman_retweet, aamir_retweet]

plt.bar(y_pos, perform, align='center', alpha=0.5)
plt.xticks(y_position, objs)
plt.ylabel('Number of retweets')
plt.title('Retweet Count Comparision ')

plt.show()
```

1) From the above bar graph, we can conclude that @BeingSalmanKhan has done more retweets than @BCCI and @aamir_khan for the fourth tweet of each account.
